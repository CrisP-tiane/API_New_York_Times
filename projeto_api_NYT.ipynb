# **Análise de Texto (Text Mining)**

A análise de texto em Python é uma técnica de processamento de linguagem natural que envolve a aplicação de algoritmos e técnicas computacionais para extrair insights e informações de dados de texto. Essa técnica é útil para explorar grandes conjuntos de dados de texto, como tweets, artigos de notícias, revisões de produtos, transcrições de conversas, entre outros.

## As principais aplicações da análise de texto em Python incluem:

**Análise de sentimento**: a identificação e classificação da emoção expressa em um texto.

**Classificação de texto**: a categorização de um texto em uma ou mais categorias, com base em seu conteúdo.

**Extração de informações**: a extração de informações relevantes de um texto, como nomes, datas, locais, entre outros.

**Análise de tópicos**: a identificação dos tópicos mais comuns em um conjunto de textos.

**Sumarização de texto**: a criação de uma versão resumida de um texto longo.

## As metodologias para análise de texto em Python podem incluir:

**Pré-processamento de texto**: a limpeza e preparação dos dados de texto, incluindo a remoção de stopwords, pontuação e outras informações irrelevantes.

**Tokenização**: a divisão do texto em unidades menores, como palavras ou frases.

**Análise morfológica**: a identificação da estrutura gramatical do texto, como verbos, substantivos e adjetivos.

**Modelagem de tópicos**: a identificação dos tópicos mais comuns em um conjunto de textos usando algoritmos de aprendizado de máquina, como LDA (Latent Dirichlet Allocation).

**Classificação de texto**: a classificação de um texto em uma ou mais categorias com base em seu conteúdo usando algoritmos de aprendizado de máquina, como Naive Bayes ou SVM (Support Vector Machines).

## Algumas das funcionalidades disponíveis em bibliotecas populares de análise de texto em Python incluem:

**NLTK (Natural Language Toolkit)**: uma biblioteca Python para processamento de linguagem natural que inclui recursos para tokenização, lematização, análise de sentimento e outras funcionalidades.

**spaCy**: uma biblioteca Python de processamento de linguagem natural que inclui recursos para análise morfológica, reconhecimento de entidades nomeadas e outras funcionalidades.

**Scikit-learn**: uma biblioteca Python de aprendizado de máquina que inclui recursos para classificação de texto, modelagem de tópicos e outras funcionalidades.

```python
# Chave de API
api_key = 'RDFBOgsmMnFO0Xi2908U73jbvGH0g8QG' # Chave do API André
```

```python
# Parâmetros para a solicitação de notícias mais populares
base_url = "https://api.nytimes.com/svc/mostpopular/v2/viewed/1.json"
params = {"api-key": api_key}
```

```python
# Fazendo a solicitação de notícias mais populares à API
response = requests.get(base_url, params=params)
data = response.json()
```

```python
# Criando uma lista para armazenar as informações relevantes de cada notícia
news = []
```

```python
# Extrair informações relevantes de cada notícia
for article in data["results"]:
    title = article["title"]
    url = article["url"]
    news.append({"title": title, "url": url})
```

```python
# Apresentando os resultados
for article in news:
    print(article["title"])
    print(article["url"])
    print()
```
